```python
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import PeftModel

# Specify the base model identifier (this should be the same one you used for training)
base_model_id = "deepseek-r1-1.5B-qwen-distill-finetuned"

# Load the tokenizer from the base model
tokenizer = AutoTokenizer.from_pretrained(base_model_id)

# Load the base model
model = AutoModelForCausalLM.from_pretrained(base_model_id, trust_remote_code=True)

# Load the LoRA adapter weights that were saved in the fine-tuned directory
model = PeftModel.from_pretrained(model, "./fine-tuned-deepseek")

# Set the model to evaluation mode
model.eval()

# Now, you can run inference
prompt = "Instruction: What is the capital of France?\nResponse:"
inputs = tokenizer(prompt, return_tensors="pt")

# Generate text (adjust max_new_tokens and other parameters as needed)
outputs = model.generate(**inputs, max_new_tokens=50)

# Decode and print the generated text
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
